```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import config
```

```python
from fredapi import Fred
fred = Fred(api_key=config.api_key)
df = {}
df['SPX'] = fred.get_series('SP500')
df['10YR'] = fred.get_series('DGS10')
data = pd.DataFrame(df)

data.tail()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
ax = data.SPX.plot() 
ax2 = ax.twinx()
ax2.plot(data['10YR'], color='orange')
ax.legend()
ax2.legend()
```
